```python
import pandas as pd
import xml.etree.ElementTree as ET
import os

# El archivo ya ha sido subido, se obtiene el nombre del archivo
input_file_path = list(uploaded.keys())[0]

data = pd.read_csv(input_file_path, encoding='latin1')

# Filtra las filas donde la columna 'STATE' esté vacía o nula
filtered_data = data[data['STATE'].isnull() | (data['STATE'].str.strip() == '')]

# Crear un directorio para guardar el archivo XML
output_directory = '/content/processed_data/'
os.makedirs(output_directory, exist_ok=True)

#Crear la ruta para el archivo XML de salida
output_file_path = os.path.join(output_directory, 'filtered_data_empty_state.xml')

# Crear la estructura XML
root = ET.Element("Data")

for _, row in filtered_data.iterrows():
    row_element = ET.SubElement(root, "Row")
    for col, value in row.items():
        col_element = ET.SubElement(row_element, col)
        col_element.text = str(value)

# Guardar el archivo XML
tree = ET.ElementTree(root)
tree.write(output_file_path)

print(f"Archivo procesado guardado en: {output_file_path}")

#Descargar el archivo generado
from google.colab import files
files.download(output_file_path)

```
